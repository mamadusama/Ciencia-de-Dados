# Pandas b√°sico

```python
import pandas as pd

BASE = "dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)

df.head()
```

```python
df.info()
```

```python
df["mfr"].value_counts().plot(kind="bar")
```

```python
# gr√°fico de contagem mfr
df["mfr"].value_counts().plot(
    kind="bar",
    xlabel="Fabricante",
    ylabel="Quantidade",
    title="Contagem de fabricantes",
)
```

```python
df["mfr"].value_counts().plot(
    kind="pie",
    ylabel="Quantidade",
    title="Percentual de fabricantes",
    autopct="%.1f%%",
)
```

```python
# m√©dia de calorias por fabricante
df.groupby("mfr")["calories"].mean().plot(
    kind="bar",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="M√©dia de calorias por fabricante",
)
```

```python
df.groupby("mfr")["calories"].mean().plot(
    kind="line",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="M√©dia de calorias por fabricante",
)
```
