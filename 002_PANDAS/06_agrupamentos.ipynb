```python
import pandas as pd

BASE = "dados/cereal.csv"

df = pd.read_csv(BASE)

df.head()
```

```python
df.groupby("mfr")["rating"].mean()
```

```python
df.groupby("mfr")[["calories", "rating"]].mean()
```

```python
df.groupby("mfr")[["calories", "rating"]].mean().sort_values(by="rating")
```

```python
df.groupby("mfr")["calories"].describe()
```

```python
df.groupby(["mfr", "type"])["calories"].describe()
```

```python
df.groupby(["mfr", "vitamins"])["calories"].describe()
```

```python
df.groupby(["mfr", "shelf"])["calories"].describe()
```
